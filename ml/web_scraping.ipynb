```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://mahadbt.maharashtra.gov.in/SchemeData/SchemeData?str=E9DDFA703C38E51AA07C7E01997E4885"
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
webpage = requests.get(url,headers=headers).text
soup = BeautifulSoup(webpage, 'html.parser')
```

```python
print(soup.prettify())
```

```python
all_categories = soup.find_all('div', class_="list resp-tab-content hor_1 resp-tab-content-active")
all_categories
```

```python
stem_url = "https://mahadbt.maharashtra.gov.in/"
```

```python
cat_list = []
href_list = []

all_categories = soup.find('div', class_="resp-tabs-container hor_1")
sub_categories_text = all_categories.find_all('li')
for i in sub_categories_text:
    cat_list.append(i.text)

anchor_tags = all_categories.find_all('a')
for i in anchor_tags:
    href_list.append(stem_url + i.get('href'))

```

```python
href_list[4]
```

```python
eligibility_list = []
for i in href_list:
    webpage = requests.get(i, headers=headers).text
    soup = BeautifulSoup(webpage, 'html.parser')
    eligibility_list.append(soup.find_all('ul', class_="list")[2].text)
```

```python
for i in eligibility_list:
    print(i)
```

```python
cleaned_eligibility_list = []

for i in eligibility_list:
    cleaned_text = i.replace('\r\n', '').strip()
    cleaned_eligibility_list.append(cleaned_text)
```

```python
eligibility_list[0]
```

```python
import pandas as pd


# Create a DataFrame from the lists
df = pd.DataFrame({
    'Scheme_name': cat_list,
    'Link': href_list,
    'Eligibilty': cleaned_eligibility_list
})

df.head()
```

```python
# Raw text
raw_text = '''
\r\n                        
\n• Student should be admitted in professional Course.
\n• Students should be the scholarship holder of Government of India  
\n• The limit of income will be applicable as per the Government of India scholarship scheme i.e. annual income limit less than or equal to 2.5 lac
\n• Students who are studying in the professional curriculum and residing in Hostel (Government. or Institutes Hostel or outside).\r\n\r\n                    
'''

# Clean the text by removing unwanted characters
cleaned_text = raw_text.replace('\r\n', '').replace('\n', '').strip()  # Remove newlines and trim

# Split the text by '•' to get each bullet point as a list item
prettified_list = [item.strip() for item in cleaned_text.split('•') if item]  # Filter out empty items

# Print each item
for item in prettified_list:
    print(item)

# Example of storing in a list
print(prettified_list)  # Outputs the prettified list
```
